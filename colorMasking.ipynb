```python
import cv2 
from google.colab.patches import cv2_imshow
import numpy as np 

image = cv2.imread('/content/retractor5.webp') 
cv2_imshow(image)
```

```python
lower_red = np.array([185, 141, 100], dtype = "uint8") 

upper_red= np.array([255,255,255], dtype = "uint8")
```

```python
mask = cv2.inRange(image, lower_red, upper_red)
```

```python
detected_output = cv2.bitwise_and(image, image, mask =  mask) 

cv2_imshow(detected_output) 

cv2.waitKey(0) 
```
